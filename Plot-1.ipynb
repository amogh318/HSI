```python
import numpy as np
#import os 
#import glob
import math
from ccdEvent4 import*
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
file4 = open('mu_tabulated_data.dat','r')
```

```python
colA= []
colB = []
for xx in file4:
    yy = xx.split()
    colA.append(float(yy[1]))
    colB.append(float(yy[5]))
```

```python
_ = plt.hist(colA,bins=180,range=(-360,360),log=True)
```

```python
_ = plt.hist(colB,bins=180,range=(-360,360),log=True)
```

```python
plt.scatter(colB,colA)
```

```python
x_edges = np.arange(0,90,5)
y_edges = np.arange(0,180,5)
```

```python
bin_values,_,__ = np.histogram2d(colB,colA,bins=(x_edges, y_edges) )
X, Y = np.meshgrid(x_edges,y_edges)
```

```python
X
```

```python
plt.pcolormesh(X, Y, bin_values.T, cmap = "rainbow", norm=colors.LogNorm() )
cbar = plt.colorbar(orientation="vertical")
plt.xlabel("ThetaTrue")
plt.ylabel("MeanTheta")
plt.savefig('MeanTheta_Vs_ThetaTrue')
```
